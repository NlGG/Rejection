```python
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
class Net(nn.Module):
    def __init__(self, num_rej=0):
        super(Net, self).__init__()
        self.num_rej = num_rej+1
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10+self.num_rej)
        self.loss = lambda g: -F.logsigmoid(-g)

    def forward(self, x):
        h = self.pool(F.relu(self.conv1(x)))
        h = self.pool(F.relu(self.conv2(h)))
        h = h.view(-1, 16 * 5 * 5)
        h = F.relu(self.fc1(h))
        h = F.relu(self.fc2(h))
        h = self.fc3(h)
        h1 = h[:, :-self.num_rej]
        h2 = h[:, -self.num_rej:]
        return h, h1, h2
    
    def pconf(self, hidden, t, loss, rate_all=1., rate_inv=0.8):
        for i in range(self.num_rej):
            if i == 0:
                h = hidden[:, i]
                loss += torch.mean(self.loss(h) + ((1-rate_all)/rate_all)*self.loss(-h))
            else:
                h = hidden[:, i]
                temp = (t == i).type(torch.FloatTensor)
                if type(rate_inv) == float:
                    temp *= rate_inv
                else:
                    temp *= rate_inv[i]

                loss += torch.mean((self.loss(h) + ((1-rate_inv)/rate_inv)*self.loss(-h))*temp)
        return loss
```

```python
for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data

        # wrap them in Variable
        inputs, labels = Variable(inputs), Variable(labels)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs, hidden1, hidden2 = net(inputs)
        #print(hidden1)
        #print(hidden2)
        loss = criterion(hidden1, labels)
        loss = net.pconf(hidden2, labels, loss, rate_all=0.3, rate_inv=0.8)
        #print(loss.data[0])
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.data[0]
        #print(loss.data[0])
        
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0

print('Finished Training')
```

```python
class Net(nn.Module):
    def __init__(self, num_rej=0):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)
        self.num_rej = num_rej+1
        self.loss = lambda g: F.logsigmoid(g)

    def forward(self, x, t, rate_all=1., rate_inv=0.8):
        print("x", x)
        h = self.pool(F.relu(self.conv1(x)))
        h = self.pool(F.relu(self.conv2(h)))
        h = h.view(-1, 16 * 5 * 5)
        h = F.relu(self.fc1(h))
        h = F.relu(self.fc2(h))
        o = F.log_softmax(self.fc3(h))
        print("o", o)
        for i in range(self.num_rej):
            if i == 0:
                print("o", o)
                o += self.loss(h) + ((1-rate_all)/rate_inv)*self.loss(-h)
            else:
                temp = t == i
                o += (self.loss(h) + ((1-rate_inv)/rate_inv)*self.loss(-h))*temp
        return o
```

```python
for epoch in range(2):  # loop over the dataset multiple times

    #
    running_loss = 0.0 


    for i, data in enumerate(trainloader):

        #
        # get the inputs
        inputs, labels = data

        # Variable
        # wrap them in Variable
        inputs, labels = Variable(inputs), Variable(labels)

        # optimizer
        # zero the parameter gradients
        optimizer.zero_grad()

        #
        # forward + backward + optimize
        outputs = model(inputs, labels)

        #Cross-Entropy
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # 
        # print statistics
        running_loss += loss.data[0]
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0

print('Finished Training')
```
